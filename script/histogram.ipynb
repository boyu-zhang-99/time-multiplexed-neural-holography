```python
from PIL import Image
import matplotlib.pyplot as plt

# Load the image
image_path = "data/2d/0.png"  # Replace with your PNG file path
image = Image.open(image_path)

# Separate the R, G, B channels
red, green, blue = image.split()

# Get pixel values for each channel
red_values = list(red.getdata())
green_values = list(green.getdata())
blue_values = list(blue.getdata())

# Plot the histograms for R, G, B
plt.figure(figsize=(10, 6))

# Red channel
plt.hist(red_values, bins=256, range=(0, 256), color="red", alpha=0.5, label="Red")
# Green channel
plt.hist(green_values, bins=256, range=(0, 256), color="green", alpha=0.5, label="Green")
# Blue channel
plt.hist(blue_values, bins=256, range=(0, 256), color="blue", alpha=0.5, label="Blue")

plt.title("RGB Histogram")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
